#Cleaning missing values in 'studio' column

```python
#replace 'studio' with 'unknown'
data['studio'] = data['studio'].fillna('unknown')
#checking to see if value returns 0
data['studio'].isnull().sum()
```

#Cleaning missing value in 'domestic_gross' column

```python
#replace 'domestic_gross' value with median
data['domestic_gross'] = data['domestic_gross'].fillna(data['domestic_gross'].median())
#checking to see if value returns 0
data['domestic_gross'].isnull().sum()
```

#Cleaning misssing value in 'foreign_gross' column

```python
#replace 'foreign_gross' value with median
data['foreign_gross'] = data['foreign_gross'].fillna(data['foreign_gross'].median())
#checking to see if value returns 0
data['foreign_gross'].isnull().sum()
```

#Cleaning misssing value in 'runtime_minutes' column

```python
#replace 'runtime_minutes' value with median
data['runtime_minutes'] = data['runtime_minutes'].fillna(data['runtime_minutes'].median())
#checking to see if value returns 0
data['runtime_minutes'].isnull().sum()
```

#Cleaning misssing value in 'genre' column

```python
#replace 'genres' value with 'unknown'
data['genres'] = data['genres'].fillna('unknown')
#checking to see if value returns 0
data['genres'].isnull().sum()
```
